```python
URL = "https://scpfoundation.net/stories"
```

```python
import time
```

```python
def get_url(URL):
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, "lxml")
  results = soup.find(class_="yui-content").find_all("a")
  item_url=[]
  for item in results:
    url = "https://scpfoundation.net" + str(item.get("href"))
    item_url.append(url)
  return item_url
```

```python
def get_stories(URL_list):
  element = []
  for url in URL_list:
      time.sleep(0.5)
      try:
        page = requests.get(url)
        soup = BeautifulSoup(page.content, "lxml")
        results = soup.find(id="page-content").find_all("p")
        for job in results:
          element.append(job.text)
        element.append("END")
      except:
        pass  
  return element
```

```python

URL_list = get_url(URL)
URL_short_list = URL_list[:30]
print(URL_short_list)
res = get_stories(URL_short_list)

print(res)
data = {}
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny2")
model = AutoModel.from_pretrained("cointegrated/rubert-tiny2")
# model.cuda()  # uncomment it if you have a GPU

def embed_bert_cls(text, model, tokenizer):
    t = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**{k: v.to(model.device) for k, v in t.items()})
    embeddings = model_output.last_hidden_state[:, 0, :]
    embeddings = torch.nn.functional.normalize(embeddings)
    return embeddings[0].cpu().numpy()

print(embed_bert_cls('привет мир', model, tokenizer).shape)
```
